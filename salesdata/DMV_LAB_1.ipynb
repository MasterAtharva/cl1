```python
# 1. Load the sales data from each file format into the appropriate data structures or 
# dataframes.
data = pd.read_csv('sales-data-sample.csv')
data
```

```python
# 2. Explore the structure and content of the loaded data, identifying any inconsistencies, 
# missing values, or data quality issues.
# 3. Perform data cleaning operations, such as handling missing values, removing 
# duplicates, or correcting inconsistencies.
data.isnull().sum()
```

```python
data = data.drop(columns=['OrderProfitable','SalesaboveTarget'])
```

```python
data.isnull().sum()
data.shape
```

```python
data = data.drop_duplicates()
data
```

```python
data.loc[:, 'order_value'] = data['Quantity'] * data['Sales']
data
```

```python
data.describe()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Group data by Category and sum the Sales
category_sales = data.groupby('Category')['SalesperCustomer'].sum().reset_index()

# Create bar plot
plt.figure(figsize=(10, 6))
sns.barplot(x='Category', y='SalesperCustomer', data=category_sales)
plt.title('Sales by Category')
plt.xlabel('Category')
plt.ylabel('Total Sales')
plt.show()
```

```python
# Group data by Country and sum the Sales
country_sales = data.groupby('Country')['SalesperCustomer'].sum()

# Create pie chart
plt.figure(figsize=(10, 6))
country_sales.plot.pie(autopct='%1.1f%%', startangle=140)
plt.title('Sales Distribution by Country')
plt.ylabel('')
plt.show()
```

```python
# Create box plot
plt.figure(figsize=(12, 8))
sns.boxplot(x='Sub_Category', y='Profit', data=data)
plt.title('Profit by Sub-Category')
plt.xlabel('Sub-Category')
plt.ylabel('Profit')
plt.xticks(rotation=45)
plt.show()
```

```python
# Group data by OrderDate and sum the Sales
time_series = data.groupby('OrderDate')['SalesperCustomer'].sum().reset_index()

# Create line plot
plt.figure(figsize=(12, 6))
sns.lineplot(x='OrderDate', y='SalesperCustomer', data=time_series)
plt.title('Sales Trends over Time')
plt.xlabel('Date')
plt.ylabel('Total Sales')
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Step 1: Load Data from Different File Formats
# Load CSV file
sales_csv = pd.read_csv('sales_data.csv')

# Load Excel file
sales_excel = pd.read_excel('sales_data.xlsx')

# Load JSON file
sales_json = pd.read_json('sales_data.json')

# Step 2: Explore Data Structure and Content
# Display first few rows of each dataset
print("CSV Data Sample:\n", sales_csv.head())
print("Excel Data Sample:\n", sales_excel.head())
print("JSON Data Sample:\n", sales_json.head())

# Check for inconsistencies and missing values
print("CSV Missing Values:\n", sales_csv.isnull().sum())
print("Excel Missing Values:\n", sales_excel.isnull().sum())
print("JSON Missing Values:\n", sales_json.isnull().sum())

# Step 3: Data Cleaning
# Handle missing values (e.g., fill or drop them)
sales_csv = sales_csv.fillna(0)
sales_excel = sales_excel.fillna(0)
sales_json = sales_json.fillna(0)

# Remove duplicates
sales_csv.drop_duplicates(inplace=True)
sales_excel.drop_duplicates(inplace=True)
sales_json.drop_duplicates(inplace=True)

# Step 4: Convert Data into a Unified Format
# Standardize column names for merging
sales_csv.columns = sales_csv.columns.str.lower()
sales_excel.columns = sales_excel.columns.str.lower()
sales_json.columns = sales_json.columns.str.lower()

# Combine into a single dataframe
all_sales_data = pd.concat([sales_csv, sales_excel, sales_json], ignore_index=True)

# Step 5: Data Transformation
# Example: Derive new column for 'Order Value' by multiplying quantity by price
all_sales_data['order_value'] = all_sales_data['quantity'] * all_sales_data['price']

# Step 6: Analyze Sales Data
# Descriptive statistics
print(all_sales_data.describe())

# Aggregations
total_sales = all_sales_data['order_value'].sum()
avg_order_value = all_sales_data['order_value'].mean()
product_category_distribution = all_sales_data['product_category'].value_counts()

print("Total Sales:", total_sales)
print("Average Order Value:", avg_order_value)
print("Product Category Distribution:\n", product_category_distribution)

# Step 7: Create Visualizations
# Sales by Product Category
plt.figure(figsize=(10, 6))
sns.barplot(x=product_category_distribution.index, y=product_category_distribution.values)
plt.title("Sales by Product Category")
plt.xlabel("Product Category")
plt.ylabel("Frequency")
plt.xticks(rotation=45)
plt.show()

# Pie chart of product category distribution
plt.figure(figsize=(8, 8))
plt.pie(product_category_distribution, labels=product_category_distribution.index, autopct='%1.1f%%')
plt.title("Product Category Distribution")
plt.show()

# Box plot for order values
plt.figure(figsize=(10, 6))
sns.boxplot(data=all_sales_data, y='order_value')
plt.title("Order Value Distribution")
plt.ylabel("Order Value")
plt.show()
```
